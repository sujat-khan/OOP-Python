## 1. What is Object-Oriented Programming?

**Object-Oriented Programming (OOP)** is a programming paradigm that organizes code around **objects** rather than functions and logic. An object is a self-contained unit that bundles:

| Component | Description | Example |
|-----------|-------------|---------|
| **Attributes** (data) | Characteristics of the object | A car's `color`, `speed`, `brand` |
| **Methods** (behavior) | Actions the object can perform | A car can `accelerate()`, `brake()`, `honk()` |

### Real-World Analogy 🏗️

Think of a **blueprint** for a house:
- The **blueprint** is the **class** — it defines the structure (rooms, doors, windows)
- Each **house built** from the blueprint is an **object** (instance) — same structure, but each can have different colors, furniture, etc.

### Why OOP?

| Benefit | Description |
|---------|-------------|
| **Modularity** | Code is organized into self-contained objects |
| **Reusability** | Classes can be reused across projects |
| **Scalability** | Easy to add new features without breaking existing code |
| **Maintainability** | Changes in one object don't affect others |
| **Real-world modeling** | Objects map naturally to real-world entities |

## 2. Procedural vs Object-Oriented: A Side-by-Side Comparison

Let's model a simple shopping scenario to see the difference.

### ❌ Procedural Approach (without OOP)

````python
# Procedural approach: using plain variables and functions
# Each item needs its own set of variables — this gets messy fast!

item1_name = "Phone"
item1_price = 100
item1_quantity = 5

item2_name = "Laptop"
item2_price = 1000
item2_quantity = 3

def calculate_total(price, quantity):
    return price * quantity

print(f"{item1_name}: ${calculate_total(item1_price, item1_quantity)}")
print(f"{item2_name}: ${calculate_total(item2_price, item2_quantity)}")

# Problem: What if we need 100 items? 100 sets of variables!
````

### ✅ Object-Oriented Approach (with OOP)

Now let's see how OOP solves this elegantly:

````python
# OOP approach: define a blueprint, then create instances

class Item:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity
    
    def calculate_total(self):
        return self.price * self.quantity

# Now creating 100 items is just 100 one-liners!
item1 = Item("Phone", 100, 5)
item2 = Item("Laptop", 1000, 3)

print(f"{item1.name}: ${item1.calculate_total()}")
print(f"{item2.name}: ${item2.calculate_total()}")
````

> 💡 **Key Insight**: Don't worry about `__init__` and `self` yet — we'll cover them in detail in the next notebook. For now, focus on the *concept* that a class bundles data and behavior together.

## 3. Your First Class: Step by Step

Let's build up from the absolute basics.

### 3.1 Defining an Empty Class

The simplest possible class uses the `pass` keyword as a placeholder:

````python
# The simplest class — does nothing yet, but it's valid Python!
class Item:
    pass

# Check that it exists
print(type(Item))  # <class 'type'> — it's a type/class
````

### 3.2 Creating Objects (Instances)

An **object** (or **instance**) is created by calling the class like a function:

```
Class Name     Calling it
    │               │
    ▼               ▼
  Item      ()      →  creates a new Item object
```

````python
class Item:
    pass

# Creating instances (objects)
item1 = Item()
item2 = Item()

# Each instance is a unique object in memory
print(f"item1 is of type: {type(item1)}")
print(f"item2 is of type: {type(item2)}")
print(f"Are they the same object? {item1 is item2}")  # False — different objects!
````

### 3.3 Adding Attributes to Objects

**Attributes** are variables that belong to an object. You can add them dynamically using dot notation:

```
object . attribute_name = value
  │          │              │
  ▼          ▼              ▼
item1  .   name      =  "Phone"
```

````python
class Item:
    pass

# Create an instance
item1 = Item()

# Assign attributes dynamically
item1.name = "Phone"
item1.price = 100
item1.quantity = 5

# Access attributes
print(f"Item: {item1.name}")
print(f"Price: ${item1.price}")
print(f"Quantity: {item1.quantity}")
print(f"Total: ${item1.price * item1.quantity}")
````

> ⚠️ **Warning**: Assigning attributes dynamically like this works, but it's **not recommended** for production code. If you create another `Item` and forget to set `price`, you'll get an `AttributeError`. We'll fix this with `__init__` in the next notebook.

````python
# Demonstrating the danger of dynamic attributes
item2 = Item()
item2.name = "Laptop"
# Oops, forgot to set price!

try:
    print(item2.price)  # This will crash!
except AttributeError as e:
    print(f"❌ Error: {e}")
    print("This is why we need __init__ to enforce required attributes!")
````

## 4. Adding Methods to a Class

A **method** is a function defined inside a class. It describes a **behavior** the object can perform.

### Key Rule: The `self` Parameter

Every instance method **must** have `self` as its first parameter. `self` refers to the **specific instance** calling the method.

```python
class Item:
    def calculate_total_price(self, x, y):
        #                      ^^^^
        #          Refers to the instance calling this method
        return x * y
```

````python
class Item:
    def calculate_total_price(self, x, y):
        """Calculate the total price given price and quantity."""
        return x * y

# Create instances and assign attributes
item1 = Item()
item1.name = "Phone"
item1.price = 100
item1.quantity = 5

item2 = Item()
item2.name = "Laptop"
item2.price = 1000
item2.quantity = 3

# Call the method on each instance
print(f"{item1.name} total: ${item1.calculate_total_price(item1.price, item1.quantity)}")
print(f"{item2.name} total: ${item2.calculate_total_price(item2.price, item2.quantity)}")
````

### 🤔 Why do we pass `item1.price` and `item1.quantity` manually?

Good question! This is clunky. In the next notebook, we'll learn how `__init__` and `self` let the method **automatically** access the object's own data:

```python
# What we'll build in Notebook 02:
class Item:
    def __init__(self, name, price, quantity):
        self.name = name
        self.price = price
        self.quantity = quantity
    
    def calculate_total_price(self):
        return self.price * self.quantity  # Uses self to access its own data!
```

## 5. Understanding `self` More Deeply

`self` is not a keyword — it's a **convention**. You could name it anything, but **always use `self`** by convention.

When you call `item1.calculate_total_price(100, 5)`, Python secretly translates it to:

```python
Item.calculate_total_price(item1, 100, 5)
#                          ^^^^^
#                  Python passes the instance automatically!
```

Let's prove this:

````python
class Item:
    def who_am_i(self):
        """Demonstrate that self is the calling instance."""
        print(f"I am: {self}")
        print(f"My type: {type(self)}")

item1 = Item()
item2 = Item()

print("=== Calling via instance ===")
item1.who_am_i()

print("\n=== Calling via class (equivalent) ===")
Item.who_am_i(item1)  # Explicitly passing the instance

print("\n=== Different instance ===")
item2.who_am_i()  # Different memory address!
````

## 6. Useful Built-in Functions for Objects

Python provides several built-in functions to inspect objects:

````python
class Item:
    def calculate_total_price(self, x, y):
        return x * y

item1 = Item()
item1.name = "Phone"
item1.price = 100

# type() — Check the type of an object
print(f"type(item1): {type(item1)}")

# isinstance() — Check if an object is an instance of a class
print(f"isinstance(item1, Item): {isinstance(item1, Item)}")
print(f"isinstance(item1, str): {isinstance(item1, str)}")

# hasattr() — Check if an object has a specific attribute
print(f"hasattr(item1, 'name'): {hasattr(item1, 'name')}")
print(f"hasattr(item1, 'quantity'): {hasattr(item1, 'quantity')}")

# getattr() — Get an attribute value with a default fallback
print(f"getattr(item1, 'name'): {getattr(item1, 'name')}")
print(f"getattr(item1, 'quantity', 0): {getattr(item1, 'quantity', 0)}")

# dir() — List all attributes and methods (including inherited ones)
print(f"\nAll attributes/methods: {[attr for attr in dir(item1) if not attr.startswith('_')]}")

# __dict__ — Dictionary of the object's attributes
print(f"item1.__dict__: {item1.__dict__}")
````

## 7. Key Terminology Summary

| Term | Definition | Example |
|------|------------|---------|
| **Class** | A blueprint/template for creating objects | `class Item:` |
| **Object / Instance** | A specific realization of a class | `item1 = Item()` |
| **Attribute** | A variable belonging to an object | `item1.name = "Phone"` |
| **Method** | A function belonging to a class | `def calculate_total_price(self):` |
| **Instantiation** | The act of creating an object from a class | `item1 = Item()` |
| **`self`** | Reference to the current instance | Used inside methods |
