## Pillar 2: Abstraction

**Abstraction** means hiding complex implementation details and exposing only what's necessary.

### Real-World Analogy ☕

When you use a coffee machine:
- **What you see**: A button that says "Make Coffee"
- **What's hidden**: Water heating, pressure regulation, grinding beans, timing

You don't need to know HOW it works — just WHAT it does.

### Encapsulation vs Abstraction

| Concept | Focus | How |
|---------|-------|-----|
| **Encapsulation** | **Hiding data** | Private attributes, getters/setters |
| **Abstraction** | **Hiding complexity** | Simple interface, complex internals |

````python
# Abstraction in action: simple interface, complex internals

class CoffeeMachine:
    """A coffee machine that hides its complexity."""
    
    def __init__(self):
        self.__water_level = 100  # percent
        self.__bean_level = 100   # percent
    
    # === PUBLIC INTERFACE (what the user sees) ===
    def make_coffee(self, coffee_type: str = "espresso"):
        """The only method the user needs to know about."""
        print(f"\n☕ Making {coffee_type}...")
        if not self.__check_resources():
            return
        self.__heat_water()
        self.__grind_beans()
        self.__brew(coffee_type)
        self.__update_levels()
        print(f"✅ Your {coffee_type} is ready!")
    
    # === PRIVATE IMPLEMENTATION (hidden from user) ===
    def __check_resources(self):
        if self.__water_level < 10:
            print("❌ Not enough water! Please refill.")
            return False
        if self.__bean_level < 10:
            print("❌ Not enough beans! Please refill.")
            return False
        return True
    
    def __heat_water(self):
        print("  🔥 Heating water to 93°C...")
    
    def __grind_beans(self):
        print("  ⚙️ Grinding beans...")
    
    def __brew(self, coffee_type):
        brew_times = {"espresso": 25, "americano": 30, "latte": 35}
        time = brew_times.get(coffee_type, 25)
        print(f"  ⏱️ Brewing for {time} seconds...")
    
    def __update_levels(self):
        self.__water_level -= 15
        self.__bean_level -= 20


# The user only needs ONE method
machine = CoffeeMachine()
machine.make_coffee("latte")
machine.make_coffee("espresso")
````

### Abstract Base Classes (ABCs)

Python has a formal way to enforce abstraction using **Abstract Base Classes**. An ABC:
- **Cannot be instantiated** directly
- Defines **abstract methods** that child classes **must** implement
- Acts as a **contract/interface**

````python
from abc import ABC, abstractmethod

class Shape(ABC):
    """Abstract base class — cannot be instantiated directly."""
    
    def __init__(self, color: str = "black"):
        self.color = color
    
    @abstractmethod
    def area(self) -> float:
        """Every shape MUST implement area(). No default!"""
        pass
    
    @abstractmethod
    def perimeter(self) -> float:
        """Every shape MUST implement perimeter()."""
        pass
    
    # Non-abstract methods can have default implementations
    def describe(self):
        return f"{self.__class__.__name__}(color={self.color}, area={self.area():.2f})"


# Can't instantiate the abstract class!
try:
    s = Shape()  # ← This will fail
except TypeError as e:
    print(f"❌ {e}")
````

````python
import math

# Concrete classes MUST implement all abstract methods

class Rectangle(Shape):
    def __init__(self, width: float, height: float, color: str = "black"):
        super().__init__(color)
        self.width = width
        self.height = height
    
    def area(self) -> float:
        return self.width * self.height
    
    def perimeter(self) -> float:
        return 2 * (self.width + self.height)


class Circle(Shape):
    def __init__(self, radius: float, color: str = "black"):
        super().__init__(color)
        self.radius = radius
    
    def area(self) -> float:
        return math.pi * self.radius ** 2
    
    def perimeter(self) -> float:
        return 2 * math.pi * self.radius


class Triangle(Shape):
    def __init__(self, a: float, b: float, c: float, color: str = "black"):
        super().__init__(color)
        self.a = a
        self.b = b
        self.c = c
    
    def area(self) -> float:
        # Heron's formula
        s = (self.a + self.b + self.c) / 2
        return math.sqrt(s * (s - self.a) * (s - self.b) * (s - self.c))
    
    def perimeter(self) -> float:
        return self.a + self.b + self.c


# All shapes can be used the same way!
shapes = [
    Rectangle(10, 5, "blue"),
    Circle(7, "red"),
    Triangle(3, 4, 5, "green"),
]

for shape in shapes:
    print(f"{shape.describe()} | Perimeter: {shape.perimeter():.2f}")
````

````python
# What if you forget to implement an abstract method?
class IncompleteShape(Shape):
    def area(self):
        return 0
    # Forgot perimeter()!

try:
    broken = IncompleteShape()
except TypeError as e:
    print(f"❌ {e}")
    print("You MUST implement ALL abstract methods!")
````

## Pillar 4: Polymorphism

**Polymorphism** (Greek: "many forms") means the same interface can behave differently depending on the object type.

### Real-World Analogy 🎮

The "+" button on a calculator:
- Numbers: 2 + 3 = 5 (addition)
- Strings: "Hello" + " World" = "Hello World" (concatenation)
- Lists: [1, 2] + [3, 4] = [1, 2, 3, 4] (merging)

Same operator, different behavior — that's polymorphism!

````python
# Built-in polymorphism with len()
print(f"len('Hello'):  {len('Hello')}")       # String: counts characters
print(f"len([1,2,3]):  {len([1, 2, 3])}")     # List: counts elements
print(f"len({{'a':1}}):  {len({'a': 1})}")     # Dict: counts keys

# Same function name, different types, different behavior!
````

### Method Polymorphism (Method Overriding)

The most common form in OOP — different classes implement the same method differently:

````python
class Animal:
    def __init__(self, name: str):
        self.name = name
    
    def speak(self):
        return "..."

class Dog(Animal):
    def speak(self):
        return f"{self.name}: Woof! 🐕"

class Cat(Animal):
    def speak(self):
        return f"{self.name}: Meow! 🐱"

class Duck(Animal):
    def speak(self):
        return f"{self.name}: Quack! 🦆"

class Snake(Animal):
    def speak(self):
        return f"{self.name}: Hiss! 🐍"


# Polymorphism: same method name, different behavior
# This function works with ANY animal — it doesn't care about the specific type
def animal_chorus(animals: list):
    """Make all animals speak — works with ANY Animal subclass."""
    print("🎵 Animal Chorus:")
    for animal in animals:
        print(f"  {animal.speak()}")  # Same method, different output!


animals = [Dog("Buddy"), Cat("Whiskers"), Duck("Donald"), Snake("Slinky")]
animal_chorus(animals)
````

### Duck Typing 🦆

Python follows the principle: **"If it walks like a duck and quacks like a duck, it's a duck."**

This means Python doesn't require inheritance for polymorphism — any object with the right methods will work:

````python
# These classes have NO common parent — but they all have speak()

class Dog:
    def speak(self):
        return "Woof!"

class Cat:
    def speak(self):
        return "Meow!"

class Robot:
    def speak(self):
        return "Beep boop!"

class Parrot:
    def speak(self):
        return "Polly wants a cracker!"


# This function doesn't care about types — just that they have speak()
def make_them_speak(things):
    for thing in things:
        print(f"  {type(thing).__name__}: {thing.speak()}")

print("Duck typing in action:")
make_them_speak([Dog(), Cat(), Robot(), Parrot()])
# No shared base class needed!
````

### Operator Overloading (Dunder Methods)

You can make your custom objects work with Python operators by implementing **dunder methods**:

````python
class Vector:
    """A 2D vector that supports mathematical operations."""
    
    def __init__(self, x: float, y: float):
        self.x = x
        self.y = y
    
    def __add__(self, other):
        """v1 + v2"""
        return Vector(self.x + other.x, self.y + other.y)
    
    def __sub__(self, other):
        """v1 - v2"""
        return Vector(self.x - other.x, self.y - other.y)
    
    def __mul__(self, scalar):
        """v * 3"""
        return Vector(self.x * scalar, self.y * scalar)
    
    def __eq__(self, other):
        """v1 == v2"""
        return self.x == other.x and self.y == other.y
    
    def __len__(self):
        """len(v) — returns magnitude as integer"""
        return int((self.x ** 2 + self.y ** 2) ** 0.5)
    
    def __repr__(self):
        return f"Vector({self.x}, {self.y})"


v1 = Vector(3, 4)
v2 = Vector(1, 2)

print(f"v1 = {v1}")
print(f"v2 = {v2}")
print(f"v1 + v2 = {v1 + v2}")
print(f"v1 - v2 = {v1 - v2}")
print(f"v1 * 3 = {v1 * 3}")
print(f"v1 == v2 = {v1 == v2}")
print(f"v1 == Vector(3, 4) = {v1 == Vector(3, 4)}")
print(f"len(v1) = {len(v1)}")
````

### Common Dunder Methods Reference

| Dunder Method | Operator/Function | Example |
|---------------|------------------|---------|
| `__init__` | Constructor | `obj = MyClass()` |
| `__repr__` | `repr()`, debugging | `print(obj)` |
| `__str__` | `str()`, user display | `str(obj)` |
| `__add__` | `+` | `obj1 + obj2` |
| `__sub__` | `-` | `obj1 - obj2` |
| `__mul__` | `*` | `obj * 3` |
| `__eq__` | `==` | `obj1 == obj2` |
| `__lt__` | `<` | `obj1 < obj2` |
| `__len__` | `len()` | `len(obj)` |
| `__getitem__` | `[]` | `obj[0]` |
| `__contains__` | `in` | `item in obj` |
| `__call__` | `()` | `obj()` |

## 5. All Four Pillars Together: A Complete Example

Let's build a **Payment Processing System** that uses all four pillars:

````python
from abc import ABC, abstractmethod
from datetime import datetime


# === ABSTRACTION: Abstract base class defines the interface ===
class PaymentProcessor(ABC):
    """Abstract base class — defines WHAT all processors must do."""
    
    def __init__(self, name: str):
        self._name = name
        self._transactions = []
    
    @abstractmethod
    def process_payment(self, amount: float) -> bool:
        """Process a payment. Must be implemented by subclasses."""
        pass
    
    @abstractmethod
    def refund(self, amount: float) -> bool:
        """Process a refund. Must be implemented by subclasses."""
        pass
    
    # Non-abstract: shared implementation
    def get_transaction_history(self):
        return self._transactions.copy()


# === INHERITANCE + ENCAPSULATION: Concrete implementations ===
class CreditCardProcessor(PaymentProcessor):
    """Processes credit card payments."""
    
    fee_rate = 0.029  # 2.9% processing fee
    
    def __init__(self, card_number: str):
        super().__init__("Credit Card")
        self.__card_number = card_number  # ENCAPSULATION: private
    
    @property
    def card_display(self):
        """Show only last 4 digits — ENCAPSULATION."""
        return f"****-****-****-{self.__card_number[-4:]}"
    
    def process_payment(self, amount: float) -> bool:
        fee = amount * self.fee_rate
        total = amount + fee
        self._transactions.append({
            "type": "payment",
            "amount": amount,
            "fee": round(fee, 2),
            "total": round(total, 2),
            "card": self.card_display,
            "time": datetime.now().strftime("%H:%M:%S")
        })
        print(f"💳 Credit Card payment: ${amount} + ${fee:.2f} fee = ${total:.2f}")
        return True
    
    def refund(self, amount: float) -> bool:
        self._transactions.append({"type": "refund", "amount": amount})
        print(f"💳 Refund of ${amount} to {self.card_display}")
        return True


class PayPalProcessor(PaymentProcessor):
    """Processes PayPal payments."""
    
    fee_rate = 0.034  # 3.4% + $0.30
    
    def __init__(self, email: str):
        super().__init__("PayPal")
        self.__email = email
    
    def process_payment(self, amount: float) -> bool:
        fee = amount * self.fee_rate + 0.30
        self._transactions.append({"type": "payment", "amount": amount, "fee": round(fee, 2)})
        print(f"🅿️ PayPal payment: ${amount} + ${fee:.2f} fee")
        return True
    
    def refund(self, amount: float) -> bool:
        self._transactions.append({"type": "refund", "amount": amount})
        print(f"🅿️ Refund of ${amount} to PayPal")
        return True


class CryptoProcessor(PaymentProcessor):
    """Processes cryptocurrency payments."""
    
    fee_rate = 0.01  # 1% fee
    
    def __init__(self, wallet_address: str):
        super().__init__("Crypto")
        self.__wallet = wallet_address
    
    def process_payment(self, amount: float) -> bool:
        fee = amount * self.fee_rate
        self._transactions.append({"type": "payment", "amount": amount, "fee": round(fee, 2)})
        print(f"₿ Crypto payment: ${amount} + ${fee:.2f} fee")
        return True
    
    def refund(self, amount: float) -> bool:
        print(f"₿ Crypto refunds are not supported!")
        return False


# === POLYMORPHISM: Same interface, different behavior ===
def checkout(processor: PaymentProcessor, amount: float):
    """Process a payment using ANY payment processor.
    
    This function doesn't know or care which specific processor it is!
    It just knows it has process_payment() — that's polymorphism.
    """
    print(f"\n{'='*50}")
    print(f"Processing ${amount} via {processor._name}...")
    success = processor.process_payment(amount)
    if success:
        print("✅ Payment successful!")
    else:
        print("❌ Payment failed!")
    return success


# Create different processors
cc = CreditCardProcessor("4532015112830366")
pp = PayPalProcessor("user@email.com")
crypto = CryptoProcessor("0x742d35Cc6634C0532925a3b844Bc9e7595f2bD")

# POLYMORPHISM: Same function, different processors
checkout(cc, 99.99)
checkout(pp, 49.99)
checkout(crypto, 199.99)

# Transaction history (encapsulated)
print(f"\nCredit Card transactions: {len(cc.get_transaction_history())}")
print(f"Card display: {cc.card_display}")  # Only shows last 4 digits
````

### ✅ Solutions

````python
# Solution 1: Notification System
from abc import ABC, abstractmethod

class Notifier(ABC):
    @abstractmethod
    def send(self, message: str):
        pass

class EmailNotifier(Notifier):
    def __init__(self, email: str):
        self.email = email
    
    def send(self, message: str):
        print(f"📧 Email to {self.email}: {message}")

class SMSNotifier(Notifier):
    MAX_LENGTH = 160
    
    def __init__(self, phone: str):
        self.phone = phone
    
    def send(self, message: str):
        if len(message) > self.MAX_LENGTH:
            message = message[:self.MAX_LENGTH - 3] + "..."
        print(f"📱 SMS to {self.phone}: {message}")

class PushNotifier(Notifier):
    def __init__(self, device_id: str):
        self.device_id = device_id
    
    def send(self, message: str):
        print(f"🔔 Push to {self.device_id}: {message}")

# Polymorphism
def notify_all(notifiers: list, message: str):
    for notifier in notifiers:
        notifier.send(message)

notifiers = [
    EmailNotifier("user@example.com"),
    SMSNotifier("+1-555-0123"),
    PushNotifier("device-abc-123"),
]

notify_all(notifiers, "Your order has been shipped!")
````

````python
# Solution 2: Money class
class Money:
    def __init__(self, amount: float, currency: str = "USD"):
        self.amount = amount
        self.currency = currency
    
    def _check_currency(self, other):
        if self.currency != other.currency:
            raise ValueError(f"Cannot operate on {self.currency} and {other.currency}")
    
    def __add__(self, other):
        self._check_currency(other)
        return Money(self.amount + other.amount, self.currency)
    
    def __sub__(self, other):
        self._check_currency(other)
        return Money(self.amount - other.amount, self.currency)
    
    def __mul__(self, scalar):
        return Money(self.amount * scalar, self.currency)
    
    def __eq__(self, other):
        return self.amount == other.amount and self.currency == other.currency
    
    def __lt__(self, other):
        self._check_currency(other)
        return self.amount < other.amount
    
    def __repr__(self):
        return f"Money({self.amount}, '{self.currency}')"
    
    def __str__(self):
        symbols = {"USD": "$", "EUR": "€", "GBP": "£"}
        symbol = symbols.get(self.currency, self.currency)
        return f"{symbol}{self.amount:.2f}"

a = Money(100, "USD")
b = Money(50, "USD")

print(f"{a} + {b} = {a + b}")
print(f"{a} - {b} = {a - b}")
print(f"{a} * 3 = {a * 3}")
print(f"{a} == {b}: {a == b}")
print(f"{b} < {a}: {b < a}")
````
